```python
# A partir desse ponto, ficarão reservadas as colunas úteis
# colunas excluídas
excluded_columns = ['id_estudante', 'id_curso']

# Colunas filtradas
std_cols_filtered = [col for col in students_all_columns if col not in excluded_columns]
```

#### 2.3.1 Medidas de posição

> Para uma rápida leitura comparativa em tabela, ficará salvo um DataFrame com as informações do método 'describe' mais a mediana, moda e as frequências de cada valor por coluna do dataset filtrado.

> Na tabela describe confirmamos que os valores batem com o que foi apresentado na introdução e na construção do dicionário de referência. Se desconsiderarmos 'id_curso' os valores variam de 0 a 7.

```python
# aqui é salvo uma cópia do describe que irá crescer com outras informações
# as colunas são invertidas com os índices para facilitar a leitura comparativa na horizontal
full_describe = describe.loc['mean':'max', :].drop('id_curso', axis=1).T
# confere exibição do full_describe
full_describe.head(3)
```

Para fazer uma limpeza visual, a partir da prévia, farei alguns tratamentos antes de adicionar os demais valores.

```python
# aqui foram arredondados para 2 casa decimais as colunas média e desvio padrão
# e retornados para inteiro os demais valores, visto que não há floats no describe original
full_describe['mean'] = full_describe['mean'].round(2)
full_describe['std'] = full_describe['std'].round(2)
full_describe['min'] = full_describe['min'].astype(int)
full_describe['25%'] = full_describe['25%'].astype(int)
full_describe['50%'] = full_describe['50%'].astype(int)
full_describe['75%'] = full_describe['75%'].astype(int)
full_describe['max'] = full_describe['max'].astype(int)
```

```python
# confere full_describe com dados tratados
full_describe.tail(3)
```

```python
# iterando sobre os valores possíveis e calculando suas frequências
# cria um dicionário que será usado na criação do dataframe
freq_table = {}
for category in np.arange(0, 8):
    # list que contera as frequências do valor iterado por colunas
    values_by_freq = []
    for col in std_cols_filtered:
      # busca no dataset original
      sum = (students_df[col] == category).sum()
      values_by_freq.append(sum)
    
    # adiciona a lista de frequência ao valor iterado no dicionário
    freq_table[category]=values_by_freq


# cria-se o dataframe que ficará de referência
freq_df = pd.DataFrame(freq_table, index=full_describe.index)
freq_df.head(3)
```

> Nessas análises as colunas/atributos assumem a posição dos índices, pois os valores representam posições e frequências de um montante de 145 registros. Foi escolhida a inversão para facilitar análise comparativa por atributo.

##### Média, Mediana, Moda e Quartis


Como a média e os quartis já existem no full_describe, o próximo passo será apresentar os demais dados e unificá-los no dataframe.

```python
# Cria-se uma série com o método 'median' a partir do dataframe original excluindo as colunas não usadas
stds_median = students_df.drop(['id_estudante','id_curso'], axis=1).median()
#atribui nome a série para facilitar na identificação quando for concatenar
stds_median.name='mediana'
```

```python
# como o processo é silimar ao da mediana, pouparei os comentários
stds_mode = students_df.drop(['id_estudante','id_curso'], axis=1).mode()
# o retorno de mode é um dataframe, por isso usarei a primeira linha como série
stds_mode = stds_mode.iloc[0].squeeze()
stds_mode.name='moda'
```

Agora serão unificados os dataframes describe, freq_df com as séries stds_median e stds_mode.

```python
# unificando através do método 'concat'
full_describe = pd.concat([stds_mode, stds_median, full_describe, freq_df], axis=1)
full_describe.tail(3)
```

##### Gráficos

As próximas células foram agrupadas em um sub-item no índice para que pudessem ser minimizadas, evitando atrapalhar a leitura do documento com grandes plotagens.

Somente alguns atributos foram plotados para não estender o arquivo com muitas imagens. Previamente foram todas renderizadas e as que chamaram minha atenção foram mantidas. Outras foram mantidas por exemplos.

Uma função foi desenvolvida para que pudesse ser replicada plotagem dos 5 gráficos.

A intenção é visualizar melhor as frequências ao redor da média, mediana e moda nos gráficos de linha, barras e pizza.

Para dar maior dimensão da frequência irei usar no gráfico de linhas os valores de forma ordenada. Pois as pontuações não seguem período de tempo ou outra marcação de ordem relevante. Assim, as instâncias dos estudantes ficarão no eixo X e os valores nos eixo Y. Busca-se uma imagem similar ao histograma porém com marcação da linha média.

Para buscar informação sobre a distribuição dos dados foi usado um gráfico de barras com a média, mediana e moda, identificando simetria ou assimetria.

```python
def plot_position_graphic(col):
    # índice dos estudantes
    x = students_df.index
    # dados da coluna na iteração em ordem crescente 
    y = students_df[col].sort_values().values

    # média, mediana e mode já obtidas
    mean = full_describe.loc[col,('Medidas de posição', 'mean')]
    median = full_describe.loc[col, ('Medidas de posição', 'mediana')]
    mode = full_describe.loc[col,('Medidas de posição', 'moda')]

    # área para duas figuras na horizontal e ajustadas
    fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3,2, figsize=(12, 9), layout='constrained')

    # atribuindo valor ao gráfico
    ax1.plot(x, y, c='silver', lw='2', label=dict_labels[col][0])
    ax1.axhline(y=mean, color='slategrey', lw='2', linestyle='--', label='Média')
    ax1.axhline(y=median, color='olivedrab', lw='2', linestyle='-', label='Mediana')
    ax1.axhline(y=mode, color='indianred',lw='2', linestyle=':', label='Moda')

    # listando os valores de refência para cada coluna 
    y_values = [value for value in dict_labels[col][1].keys()]
    y_values.append(mean)

    # listando as legendas de refência para cada valor 
    y_labels = [label for label in dict_labels[col][1].values()]
    y_labels.append('')
    
    # atribuindo ao gráfico os valores e legendas
    ax1.set_yticks(y_values,y_labels)
    ax1.set_xticks([1, 145])

    # atribuindo e ajustando etiquetas e legenda
    ax1.set_ylabel(col)
    ax1.set_xlabel(f'estudante ordenado por valor do dado')
    ax1.set_title(f'Gráfico de linha com média')
    ax1.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0))

    #
    ax2.bar(['média', 'mediana', 'moda'],[mean, median, mode], color=['slategrey', 'olivedrab', 'indianred'])
    title = ''
    if mean < median < mode :
        title = 'Assimetria à esquerda'
    elif mean == median == mode:
        title = 'Simétrico'
    elif mean > median > mode:
        title = 'Assimetria à direita'
    else:
        title = 'Sem distribuição clara'
    ax2.set_title(title)


    # traz a soma das frequências dos dados acima e abaixo da média
    above_mean = int(students_df[col].gt(mean).sum())
    below_mean = int(students_df[col].lt(mean).sum())

    # mediana
    above_median = int(students_df[col].gt(median).sum())
    equal_median = int(students_df[col].eq(median).sum())
    below_median = int(students_df[col].lt(median).sum())

    # moda
    above_mode = int(students_df[col].gt(mode).sum())
    equal_mode = int(students_df[col].eq(mode).sum())
    below_mode = int(students_df[col].lt(mode).sum())
    
    # atribui os resultados acima e abaixo da média ao gráfico de pizza
    # configura legenda, estilo e título
    ax3.pie([above_mean, below_mean], labels=['Acima da média', 'Abaixo da média'], colors=['slategrey', 'silver'], autopct='%1.1f%%', startangle=90, explode=[0.05, 0])
    ax3.set_title('Total dos valores em relação a média')

    ax4.pie([above_median, equal_median, below_median], labels=['Acima da mediana','Igual a mediana', 'Abaixo da mediana'], colors=['silver', 'olivedrab', 'silver'], autopct='%1.1f%%', startangle=90, explode=[0, 0.05, 0.05])
    ax4.set_title('Total dos valores em relação a mediana')

    ax5.pie([above_mode, equal_mode, below_mode], labels=['Acima da moda','Igual a moda', 'Abaixo da moda'], colors=['silver', 'indianred', 'silver'], autopct='%1.1f%%', startangle=90, explode=[0, 0.05, 0.05])
    ax5.set_title('Total dos valores em relação a moda')


    fig.delaxes(ax6)
    # ajusta a tela e apresenta o plot com margem de linhas
    plt.tight_layout()
    plt.show()
    print("\n\n\n")
```

#### 2.3.2 Medidas de dispersão

##### Amplitude, Variância, Desvio padrão e Coeficiente de Variação

> No formato que estão apresentados os dados, a leitura da amplitude amostral não nos informa nada de relevante. Uma vez que valores do dataset já estão tratados em representações, a amplitude amostral nos atributos variam de 1(valores binários de sim e não) a 7(valores das notas que, depois de id_curso, tem maior variação).

> Ainda que se transcrevesse nos atributos que tratam de alcance de valores, não poderia se determinar valor final registrado. Em 'idade' os dados representam de 18 ao indefinido e 'salario' de 135 ao indefinido.

```python
# criando série de variações a partir do dataframe original com colunas filtradas
var = students_df.drop(['id_estudante', 'id_curso'], axis=1).var()
# puxando valores de desvio padrão do full_describe
std = full_describe.loc[:, ('Medidas de posição', 'std')]
# criando série de coeficiente de variação a partir do desvio padrão e média do full_decribe
coef =( std / full_describe.loc[:, ('Medidas de posição', 'mean')] )* 100

# concatenando valores para comparação
compare_scatter = pd.concat([var, std, coef], axis=1)
# renomeando as colunas para fácil leitura
compare_scatter.columns=['variação', 'desvio padrão', 'coeficiente de var']
# imprimindo o dataframe comparativo
compare_scatter
```

##### Sobre Variância, Desvio padrão e Coeficiente de Variação

> Nota-se que para os valores das variações de dispersão, a grande maioria é próxima de zero.

> Complementando as sinalizações anteriores (nas análises das medidas de posição), 'estado_civil_pais', 'impacto_do_projeto_no_sucesso' e 'organizacao_avaliacoes' possuem baixa variação, indicando maior probabilidade de outliers nos casos que vimos a linha se dispersando da linha média.

Para agilidade, já foram impressos os dados mais simples de forma automática, facilitando as análises.

```python
# imprimindo os 3 menores e os 3 maiores valores da tabela comparativa
# usando a ordenação por coluna e retirando pelos índices
lowest_var = compare_scatter['variação'].sort_values()[:3]
biggest_var = compare_scatter['variação'].sort_values()[-3:]

# buscando a porcentagem de variação abaixo de 1.0
# para definir a porcentagem será feito em conta  sobre os atributos usados
# 
percentage = (100 / len(compare_scatter.index))
lower_percent_var = compare_scatter['variação'].lt(1.0).sum()*percentage

print(f'Os atributos que apresentam as 3 menores variações são {[attr for attr in lowest_var.index]} com {[value.round(3) for value in lowest_var.values]}')
print(f'Os atributos que apresentam as 3 maiores variações são {[attr for attr in biggest_var.index]} com {[value.round(3) for value in biggest_var.values]}')
print(f'Aproximadamente {lower_percent_var.round(2)}% dos atributos possuem variações relativas menores que 1.0.')
```

```python
# repetindo o processo como nas análises de variações, porém com o desvio padrão
lowest_std = compare_scatter['desvio padrão'].sort_values()[:3]
biggest_std = compare_scatter['desvio padrão'].sort_values()[-3:]

lower_percent = compare_scatter['desvio padrão'].lt(1.0).sum()*percentage

print(f'Os atributos que apresentam os 3 menores devios padrão são {[attr for attr in lowest_std.index]} com {[value.round(3) for value in lowest_std.values]}')
print(f'Os atributos que apresentam os 3 maiores desvios padrão são {[attr for attr in biggest_std.index]} com {[value.round(3) for value in biggest_std.values]}')
print(f'Aproximadamente {lower_percent.round(2)}% dos atributos possuem desvio padrão menor que 1.0.')
```

> Como era esperado, o desvio padrão acompanha a variância. 74% dos atributos que possuem desvio padrão abaixo de 1. Ou seja, grande concentração de dados nas colunas do dataset original.

```python
# repetindo o processo como nas análises de variações, porém com o coeficiente de var
lowest_coef = compare_scatter['coeficiente de var'].sort_values()[:3]
biggest_coef = compare_scatter['coeficiente de var'].sort_values()[-3:]

bigger_percent_coef = compare_scatter['coeficiente de var'].gt(25).sum()*percentage

print(f'Os atributos que apresentam os 3 menores devios padrão são {[attr for attr in lowest_coef.index]} com {[value.round(3) for value in lowest_coef.values]}')
print(f'Os atributos que apresentam os 3 maiores desvios padrão são {[attr for attr in biggest_coef.index]} com {[value.round(3) for value in biggest_coef.values]}')
print(f'Aproximadamente {bigger_percent_coef.round(2)}% dos atributos possuem coeficiente de var maior que 25%.')
```

> Já o coeficiente de variação nos indica que a variação relativa dos dados é média ou alta(acima dos 25%) em 93% dos atributos. Mostra-se uma alta dispersão dos dados em relação à média. Indícios de dados altamente assimétricos e/ou outliers.
